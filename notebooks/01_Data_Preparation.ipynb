## 4. Missing Value Analysis

```python
# Missing value analysis

missing_values = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Missing Percentage': round((df.isnull().sum()/len(df))*100, 2)
})

missing_values[missing_values['Missing Values'] > 0] \
    .sort_values(by='Missing Values', ascending=False)
```

### Missing Value Findings

The missing value analysis shows that only two columns contain null values:

- `request_date`
- `return_reason`

Both columns have approximately **94.48% missing values**, corresponding to **32,597 orders**.

This is expected because these fields are only populated when a customer initiates a return request. Since only **1,903 orders were returned**, non-returned orders naturally do not have a return request date or return reason.

Therefore, these missing values represent valid business behavior rather than data quality issues and will be retained without imputation.

### Decision

- No missing value treatment is required.
- `request_date` and `return_reason` will be preserved for return-specific analysis.

## 6. Date Conversion

```python
# Convert date columns using the correct format

date_columns = [
    'order_date',
    'delivered_date',
    'request_date'
]

for col in date_columns:
    df[col] = pd.to_datetime(
        df[col],
        format='%d/%m/%Y',
        errors='coerce'
    )

df[date_columns].dtypes
```

## 7. Feature Engineering: Delivery Duration

To evaluate delivery performance, a new feature called `delivery_days` is created.

The feature represents the number of days between order placement and product delivery.

```python
# Create delivery duration feature

df['delivery_days'] = (
    df['delivered_date'] - df['order_date']
).dt.days

# Summary statistics

df['delivery_days'].describe()
```

#### Interpretation

Delivery durations appear consistent across the dataset with no negative or unrealistic values observed.

Although most deliveries occur within a narrow range, longer delivery durations may influence customer satisfaction and contribute to higher return likelihood.

## 8. Feature Engineering: Customer Age Groups
Customer age is currently available as a numerical variable.

To improve interpretability and support customer segmentation analysis, age groups will be created.

```python
# Age distribution summary

df['customer_age'].describe()
```

```python
# Create age groups

bins = [18, 25, 35, 45, 60, 100]

labels = [
    '18-25',
    '26-35',
    '36-45',
    '46-60',
    '60+'
]

df['age_group'] = pd.cut(
    df['customer_age'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Verify distribution

df['age_group'].value_counts().sort_index()
```

#### Interpretation
The customer base is distributed across all age groups, with the 46–60 segment representing the largest share of customers.

The resulting age groups provide a more interpretable representation of customer demographics and will support return behavior analysis across different customer segments.

## 9. Target Variable Preparation

The `returned` column currently contains categorical values:

- Yes
- No

For machine learning purposes, the target variable will be encoded as a binary field:

| Original Value | Encoded Value |
|---------------|--------------|
| Yes | 1 |
| No | 0 |

This transformation enables the use of classification algorithms while preserving the original business meaning.

```python
# Examine target distribution before encoding

df['returned'].value_counts()
```

#### Return Rate

Overall Return Rate ≈ 5.52%

#### Interpretation

Only a small percentage of orders are returned, indicating a significant class imbalance.

This is an important consideration for predictive modeling because a model that predicts every order as "Not Returned" would achieve high accuracy while providing little business value.

```python
# Encode target variable

df['returned'] = df['returned'].map({
    'No': 0,
    'Yes': 1
})

# Verify conversion

df['returned'].value_counts()
```
